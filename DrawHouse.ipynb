```python
import turtle as tl
import math
screen = tl.Screen()
screen.bgcolor("light blue")
pen = tl.Turtle()
pen.pensize(3)
def draw_shapes(col,length,height,degree_1,degree_2):
    pen.color(col)
    for i in range(0,2):
        pen.forward(length)
        pen.left(degree_1)
        pen.forward(height)
        pen.left(degree_2)
# vẽ mặt tiền
pen.fillcolor("green")
pen.begin_fill()
draw_shapes("green",200,200,90,90)
pen.end_fill()
# vẽ cửa chính
pen.penup()
pen.goto(50,0)
pen.pendown()
pen.fillcolor("gray")
pen.begin_fill()
draw_shapes("gray",100,100,90,90)
pen.end_fill()
# Vẽ sườn ngôi nhà
pen.penup()
pen.goto(200,0)
pen.pendown()
pen.fillcolor("blue")
pen.begin_fill()
pen.left(30)
draw_shapes("blue",150,200,60,120)
pen.end_fill()
# Vẽ cửa sổ
pen.penup()
pen.goto(250,100)
pen.pendown()
pen.fillcolor("brown")
pen.begin_fill()
draw_shapes("brown",50,70,60,120)
pen.end_fill()
# Vẽ mái nhà
pen.penup()
pen.goto(0,200)
pen.pendown()
pen.fillcolor("yellow")
pen.begin_fill()
pen.color("yellow")
pen.right(30)
pen.forward(200)
pen.left(135)
pen.forward(200/math.sqrt(2))
pen.left(90)
pen.forward(200/math.sqrt(2))
pen.end_fill()
pen.penup()
pen.goto(200,200)
pen.pendown()
pen.fillcolor("violet")
pen.begin_fill()
pen.left(165)
draw_shapes("violet",150,200/math.sqrt(2),105,75)
pen.end_fill()
# Vẽ cây
# vẽ thân cây
pen.penup()
pen.goto(-150,-100)
pen.pendown()
pen.fillcolor("brown")
pen.begin_fill()
pen.right(120)
draw_shapes("brown",100,50,90,90)
pen.end_fill()
# Vẽ lá
pen.penup()
pen.goto(-225,-100)
pen.pendown()
pen.left(90)
pen.fillcolor("green")
pen.begin_fill()
pen.color("green")
pen.forward(200)
pen.left(135)
pen.forward(200/math.sqrt(2))
pen.left(90)
pen.forward(200/math.sqrt(2))
for i in range(1,3):
    pen.penup()
    pen.goto(-225,-100 +40*i)
    pen.pendown()
    pen.left(135)
    pen.forward(200)
    pen.left(135)
    pen.forward(200/math.sqrt(2))
    pen.left(90)
    pen.forward(200/math.sqrt(2))
pen.end_fill()
# Vẽ mặt trời
pen.fillcolor("yellow")
pen.begin_fill()
pen.color("yellow")
pen.penup()
pen.goto(-300,300)
pen.pendown()
pen.circle(100)
pen.end_fill()
tl.done()
```
